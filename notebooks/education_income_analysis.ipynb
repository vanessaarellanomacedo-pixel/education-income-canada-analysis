## Median vs Average Employment Income

This table compares median employment income and average employment income by educational attainment level in Canada in 2020.

The average employment income is generally higher than the median employment income. This suggests that some groups may include individuals with very high incomes, which increases the average. For this reason, the median employment income is useful because it better represents the typical income within each education level.

Overall, comparing both measures provides a more complete understanding of income patterns across educational attainment levels.

```python
import numpy as np

income_comparison_sorted = income_comparison.sort_values(
    by="median_employment_income",
    ascending=False
)

x = np.arange(len(income_comparison_sorted["education_level"]))
width = 0.35

plt.figure(figsize=(14, 8), facecolor="white")

plt.bar(
    x - width/2,
    income_comparison_sorted["median_employment_income"],
    width,
    label="Median Employment Income"
)

plt.bar(
    x + width/2,
    income_comparison_sorted["average_employment_income"],
    width,
    label="Average Employment Income"
)

plt.xlabel("Educational Attainment", color="black")
plt.ylabel("Employment Income ($)", color="black")
plt.title("Median vs Average Employment Income by Educational Attainment in Canada, 2020", color="black")

plt.xticks(
    x,
    income_comparison_sorted["education_level"],
    rotation=90,
    color="black"
)

plt.yticks(color="black")
plt.legend()

plt.gca().set_facecolor("white")
plt.tight_layout()

plt.savefig(
    "../images/median_vs_average_income_by_education_canada_2020.png",
    bbox_inches="tight",
    facecolor="white"
)

plt.show()
```

In all education groups, the average employment income is higher than the median employment income. This suggests that income distribution may be right-skewed, meaning that some individuals with higher earnings increase the average income.

The largest difference appears in the degree in medicine, dentistry, veterinary medicine, or optometry category, where the average employment income is much higher than the median employment income. This indicates that some individuals in this group may have significantly higher earnings compared with the typical income level.

For this reason, median employment income is useful because it better represents the typical income within each education level, while average employment income helps identify the influence of higher-income earners.
